We will start by analysing the sales over time.

```python
# Converting 'order date (DateOrders)' to datetime
data_df['order date (DateOrders)'] = pd.to_datetime(data_df['order date (DateOrders)'])

# Creating a new column for 'Year-Month'
data_df['Year_Month'] = data_df['order date (DateOrders)'].dt.to_period('M')

# Grouping sales over time by month to observe monthly sales
monthly_sales = data_df.groupby('Year_Month')['Sales per customer'].sum()
```

```python
# Plotting total sales over time
plt.figure(figsize=(20, 6))
monthly_sales.plot(title='Total Sales Over Time (Monthly)', color='blue', marker='o')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

We can observe fluctuation in sales over time but they oscillate between 800,000 and 1 million. This might be due to seasonal trends or other market dynamics. The sales went over 1 million in September, 2017 which was the highest ever.

Next, we will check the top selling products and the most profitable markets i.e. generate the most revenue.

```python
# Best-selling products
best_selling_products = data_df.groupby('Product Name')['Sales per customer'].sum().nlargest(10)

# Most profitable markets
most_profitable_markets = data_df.groupby('Market')['Sales per customer'].sum().nlargest(10)
```

```python
# Plotting best-selling products and most profitable markets
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))
best_selling_products.plot(kind='bar', ax=axes[0], color='green', title='Best-Selling Products')
axes[0].set_ylabel('Total Sales')
axes[0].set_xlabel('Product Name')

most_profitable_markets.plot(kind='bar', ax=axes[1], color='red', title='Most Profitable Markets')
axes[1].set_ylabel('Total Sales')
axes[1].set_xlabel('Market')

plt.tight_layout()
plt.show()
```

- Field & Stream Sportsman 16 Gun Fire Safe was the best selling product with over 6 million in sales.
- Europe was the most profitable market followed by LATAM with over 9 million sales each.

We will also check the product category which generates the most sales.

```python
# Analyze product categories that generate the most sales
category_sales = data_df.groupby('Category Name')['Sales per customer'].sum().sort_values(ascending=False)


print(category_sales.head(10))

plt.figure(figsize=(20,5))
sns.barplot(data=category_sales.head(10))
```

**Fishing** equipment has the highest total sale per customer among all categories. This indicates high customer interest and/or expensive items. All these categories can be used as targets for promotional campaigns or inventory adjustments based on sales performance.

Next we will analyse delivery performance.

```python
delivery_status = data_df['Delivery Status'].value_counts()

sns.barplot(delivery_status)
```

There are a lot of orders with **Late delivery**

```python
# Determine the repeat purchase rate
# Count the number of orders per customer
customer_orders = data_df['Customer Id'].value_counts()

# Customer with order more than 1 would be repeat customers
repeat_customers = customer_orders[customer_orders>1].count()

total_customers = customer_orders.count()

repeat_purchase_rate = repeat_customers/total_customers

print(f'Approximately {round(repeat_purchase_rate*100, 2)}% of customers have made more than one purchase')
```

To understand customer purchase behaviour better, we will calculate the repeat purchase rates specifically for customers who have experienced late daliveries vs those who had on-time or advanced deliveries.

```python
# Map customer IDs to their delivery experiences
customer_delivery_experience = data_df.groupby('Customer Id')['Delivery Status'].apply(lambda x: 'Late' if 'Late delivery' in x.values else 'On-time/Advanced')

# Join this with the customer_orders to see how many times each customer ordered
customer_orders_with_delivery = customer_orders.to_frame(name='Number of Orders').join(customer_delivery_experience)

# Calculate repeat purchase rates for both groups
repeat_rate_late = customer_orders_with_delivery[customer_orders_with_delivery['Delivery Status'] == 'Late']['Number of Orders'].value_counts().loc[lambda x: x.index > 1].sum() / customer_orders_with_delivery[customer_orders_with_delivery['Delivery Status'] == 'Late'].shape[0]
repeat_rate_ontime = customer_orders_with_delivery[customer_orders_with_delivery['Delivery Status'] == 'On-time/Advanced']['Number of Orders'].value_counts().loc[lambda x: x.index > 1].sum() / customer_orders_with_delivery[customer_orders_with_delivery['Delivery Status'] == 'On-time/Advanced'].shape[0]

print(f'Approximately {round(repeat_rate_late*100,2)}% of customers who experienced late deliveries have made more than one purchase.')
print(f'Approximately {round(repeat_rate_ontime*100,2)}% of customers who experienced on-time or advanced deliveries have made more than one purchase.')
```

Surprisingly, customers who experienced late deliveries have significantly higher repeat purchase rate compared to those with one-time or advanced deliveries. This might be due to other factors like discounts, product quality, customer service, etc.

Next, we will plot the Geographical distribution of sales and customers to understand the regions of sale and purchase.

```python
# Create a new figure with a specific size
plt.figure(figsize=(12, 8))

# Set up the Basemap projection
m = Basemap(projection='merc', llcrnrlat=-60, urcrnrlat=70, llcrnrlon=-180, urcrnrlon=180, lat_ts=20, resolution='c')

# Draw coastlines, countries, and states
m.drawcoastlines()
m.drawcountries()
m.drawstates()

# Convert latitude and longitude to the map projection coordinates
x, y = m(data_df['Longitude'].values, data_df['Latitude'].values)

# Plotting the data, scaling the dots by the 'Sales per customer'
scatter = m.scatter(x, y, alpha=0.6, edgecolor='none', c='blue', s=data_df['Sales per customer']/data_df['Sales per customer'].max()*100, label='Sales per Customer')

# Add a color bar
plt.colorbar(scatter, label='Sales per Customer')

# Add title and labels (not needed for longitude and latitude as it's a map)
plt.title('Geographical Distribution of Sales')

# Show the plot
plt.show()
```

- This scatter plot shows the geographical distribution of sales, where each point represents a location. The size and color indicates the magnitude of sales at the location.
- The larger and darker regions indicate higher sales, while the smaller and lighter points show lower sales.

Let's analyse the effect of discount on sales. We will observe the correlation between discount rate and sales per customer and plot the trends in sales values with varying discounts.

```python
# Calculate the correlation between discount rate and sales per customer
correlation = data_df['Order Item Discount Rate'].corr(data_df['Sales per customer'])
print(f"Correlation between Discount Rate and Sales per Customer: {correlation}")

# Scatter plot to visualize the relationship between discount rate and sales per customer
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Order Item Discount Rate', y='Sales per customer', data=data_df)
plt.title('Impact of Discount Rate on Sales per Customer')
plt.xlabel('Discount Rate')
plt.ylabel('Sales per Customer')
plt.grid(True)
plt.show()
```

The correlation between discount rate and sales per customer is ~ -0.12, indicating a slight **negative relationship**. This means as the discount rate increases, the sales per customer slightly decreases. This is a bit counterintuitive as we expect the opposite to happen!

Next, we will analyse the shipping modes and understand their distribution and impact on delivery times. We will first look at the frequency of each shipping mode.

```python
# Analyze the frequency of each shipping mode
shipping_mode_counts = data_df['Shipping Mode'].value_counts()
print(shipping_mode_counts)
sns.barplot(data=shipping_mode_counts)
```

**Standard Class** is the most commonly used shipping mode. Next, we will  check the average delivery time due to each shipping mode.

```python
# Analyze the relationship between shipping mode and average delivery times
average_delivery_times = data_df.groupby('Shipping Mode')['Days for shipping (real)'].mean()

print(average_delivery_times)

sns.barplot(data=average_delivery_times)
```

Obviously, **Same Day** shipping has the fastest delivery time, but **Second Class** and **Standard Class** take the longest to be delivered which is nearly 4 days.

Now, we will look at Delivery status by shipping mode.

```python
# Analyze the relationship between shipping mode and delivery status
shipping_delivery_status = data_df.groupby(['Shipping Mode', 'Delivery Status']).size().unstack()

shipping_delivery_status
```

```python
# Plotting
ax = shipping_delivery_status.plot(kind='bar', figsize=(10, 6), width=0.8)
plt.title('Delivery Status by Shipping Mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability

# Adding value labels on top of each bar
for p in ax.patches:
    if pd.notna(p.get_height()):
        ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.tight_layout()
```

- First Class and Second Class are more or less similar with almost the same amount of late deliveries and a few order cancellations.
- Same Day deliveries 50~50 on time or late with a few cancellations.
- Standard Class has a mix of advance shipping, highest late deliveries and On time deliveries atleast half of the late deliveries. It has nearly 4.5k cancellations as well.
